```python
# import library
import pandas as pd
from scipy.spatial.distance import cityblock
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
# Dataset
data = {
    'DTW': [2, 3, 4, 1, 3],
    'DTT': [4, 4, 3, 5, 2],
    'DMT': [4, 3, 2, 4, 1],
    'DDB': [3, 5, 5, 2, 3]
}

df = pd.DataFrame(data)
```

```python
# Fungsi jarak Manhattan
def manhattan_distance(x, y):
    return cityblock(x, y)
```

```python
# AHC dengan complete linkage
complete_linkage = linkage(df.values, method='complete', metric=manhattan_distance)
plt.title('Dendrogram Menggunakan Complete Linkage')
dendrogram(complete_linkage, labels=df.index)
plt.show()
```

```python
# AHC dengan average linkage
average_linkage = linkage(df.values, method='average', metric=manhattan_distance)
plt.title('Dendrogram Menggunakan Average Linkage')
dendrogram(average_linkage, labels=df.index)
plt.show()
```
