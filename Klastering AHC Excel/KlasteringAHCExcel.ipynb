```python
# import library
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("Data AHC.xlsx")
print(data)
```

```python
x = np.asarray(data)
print(x)
```

```python
# lakukan hierarchical klastering
Z = linkage(x)
```

```python
# plot dendrogram
plt.figure(figsize=(25, 10))
plt.title('Dendogram Hierarchical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(Z)
plt.show()
```
